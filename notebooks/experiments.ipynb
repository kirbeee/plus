# DataSet Statistics

```python
import os
from collections import Counter

# 🔴 改成你的實際根目錄
ROOT_DIR = "/Users/sean_lin/Documents/github_repo/plus/data/PLUSVein-FV3"

stats = {
    "total": 0,
    "data_type": Counter(),   # RAW / ROI / SingleFinger
    "scanner": Counter(),     # PLUS_FV3_Laser / PLUS_FV3_LED
    "side": Counter(),        # DORSAL / PALMAR
    "session": Counter(),
    "user": Counter(),
    "finger": Counter(),
}

def parse_filename(filename):
    """
    支援：
    PLUS_FV3_Laser_DORSAL_01_001_02_01.png
    PLUS-FV3-Laser_DORSAL_01_001_02_01.png
    """
    name = filename.replace(".png", "")
    parts = name.split("_")
    if len(parts) != 6:
        return None

    scanner, side, session, user, finger, image = parts
    return scanner, side, session, user, finger, image


def detect_data_type(path):
    if "RAW" in path:
        return "RAW"
    if "ROI" in path:
        return "ROI"
    if "SingleFinger" in path:
        return "SingleFinger"
    return "Unknown"


def traverse_directory(current_path):
    try:
        for entry in os.listdir(current_path):
            full_path = os.path.join(current_path, entry)

            # 如果是資料夾 → 繼續遞迴
            if os.path.isdir(full_path):
                traverse_directory(full_path)

            # 如果是 png 檔 → 分析
            elif entry.lower().endswith(".png"):
                parsed = parse_filename(entry)
                if parsed is None:
                    return

                scanner, side, session, user, finger, image = parsed
                data_type = detect_data_type(full_path)

                stats["total"] += 1
                stats["data_type"][data_type] += 1
                stats["scanner"][scanner] += 1
                stats["side"][side] += 1
                stats["session"][session] += 1
                stats["user"][user] += 1
                stats["finger"][finger] += 1

    except PermissionError:
        pass


# 🚀 開始遞迴
traverse_directory(ROOT_DIR)

# ======================
# 輸出統計結果
# ======================
print("\n========== 遞迴遍歷統計結果 ==========\n")

print(f"總影像數量: {stats['total']}\n")

print("【資料類型】")
for k, v in stats["data_type"].items():
    print(f"{k}: {v}")

print("\n【Scanner】")
for k, v in stats["scanner"].items():
    print(f"{k}: {v}")

print("\n【拍攝方向】")
for k, v in stats["side"].items():
    print(f"{k}: {v}")

print("\n【Session】")
for k, v in stats["session"].items():
    print(f"Session {k}: {v}")

print("\n【Finger 分佈】")
for k, v in sorted(stats["finger"].items()):
    print(f"Finger {k}: {v}")

print("\n【前 10 位使用者影像數】")
for user, v in sorted(stats["user"].items())[:10]:
    print(f"User {user}: {v}")
```
